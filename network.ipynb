```python
import networkx as nx
```

```python
EASE = 'Ease of Use'
CHANGE_AVERSION = 'Change Aversion'
TIME_COMMITMENT = 'Time Commitment'
FAMILIARITY = 'Organizational Familiarity'
DISCOVERABILITY = 'Information Discoverability'
DISTRIBUTION = 'Ease of Distribution'
QA = 'Easier Q&A'
CASUALNESS = 'Casual Platform'
EXPLORABILITY = 'Easy to find out how to get info'
CUSTOMER_SERVICE = 'Customer Service'
LEADERSHIP_SUPPORT = 'Endorsed by Leadership'

# EASE = 'A'
# CHANGE_AVERSION = 'B'
# TIME_COMMITMENT = 'C'
# FAMILIARITY = 'D'
# DISCOVERABILITY = 'E'
# DISTRIBUTION = 'F'
# QA = 'G'
# CASUALNESS = 'H'
# EXPLORABILITY = 'I'
# CUSTOMER_SERVICE = 'J'
# LEADERSHIP_SUPPORT = 'K'

graph = [
    (CHANGE_AVERSION, TIME_COMMITMENT, 2),
    (CHANGE_AVERSION, CHANGE_AVERSION, 2),
    (CHANGE_AVERSION, FAMILIARITY, 2),
    (CHANGE_AVERSION, CASUALNESS, 1),
    (CHANGE_AVERSION, CUSTOMER_SERVICE, 2),

    (FAMILIARITY, FAMILIARITY, 2),
    (FAMILIARITY, TIME_COMMITMENT, 1),
    (FAMILIARITY, DISCOVERABILITY, 1),
    (FAMILIARITY, CHANGE_AVERSION, 2),
    (FAMILIARITY, CASUALNESS, 1),
    (FAMILIARITY, EXPLORABILITY, 1),

    (TIME_COMMITMENT, CHANGE_AVERSION, 2),
    (TIME_COMMITMENT, DISCOVERABILITY, 1),
    (TIME_COMMITMENT, EASE, 1),
    (TIME_COMMITMENT, EXPLORABILITY, 2),
    (TIME_COMMITMENT, CUSTOMER_SERVICE, 2),

    (DISCOVERABILITY, EASE, 2),
    (DISCOVERABILITY, TIME_COMMITMENT, 1),
    (DISCOVERABILITY, CASUALNESS, 1),
    (DISCOVERABILITY, EXPLORABILITY, 1),
    (DISCOVERABILITY, CUSTOMER_SERVICE, 2),

    (EASE, TIME_COMMITMENT, 1),
    (EASE, EASE, 1),
    (EASE, CHANGE_AVERSION, 1),
    (EASE, DISCOVERABILITY, 1),
    (EASE, CASUALNESS, 1),
    (EASE, EXPLORABILITY, 1),
    (EASE, CUSTOMER_SERVICE, 1),

    (DISTRIBUTION, TIME_COMMITMENT, 2),
    (DISTRIBUTION, DISCOVERABILITY, 1),
    (DISTRIBUTION, QA, 1),
    (DISTRIBUTION, CASUALNESS, 1),
    (DISTRIBUTION, EXPLORABILITY, 1),
    (DISTRIBUTION, EASE, 1),
    (DISTRIBUTION, CUSTOMER_SERVICE, 1),

    (QA, DISCOVERABILITY, 1),
    (QA, CASUALNESS, 2),
    (QA, EXPLORABILITY, 1),
    (QA, EASE, 1),
    (QA, CHANGE_AVERSION, 1),
    (QA, CUSTOMER_SERVICE, 2),

    (CASUALNESS, DISCOVERABILITY, 2),
    (CASUALNESS, QA, 1),
    (CASUALNESS, EXPLORABILITY, 1),
    (CASUALNESS, EASE, 1),
    (CASUALNESS, CHANGE_AVERSION, 1),
    (CASUALNESS, CUSTOMER_SERVICE, 1),

    (EXPLORABILITY, TIME_COMMITMENT, 1),
    (EXPLORABILITY, DISCOVERABILITY, 1),
    (EXPLORABILITY, EASE, 1),
    (EXPLORABILITY, CUSTOMER_SERVICE, 1),

    (CUSTOMER_SERVICE, TIME_COMMITMENT, 2),
    (CUSTOMER_SERVICE, DISCOVERABILITY, 1),
    (CUSTOMER_SERVICE, EASE, 1),
    (CUSTOMER_SERVICE, CHANGE_AVERSION, 1),
    (CUSTOMER_SERVICE, FAMILIARITY, 1),

    (LEADERSHIP_SUPPORT, TIME_COMMITMENT, 1),
    (LEADERSHIP_SUPPORT, CHANGE_AVERSION, 2),
    (LEADERSHIP_SUPPORT, FAMILIARITY, 2),
    (LEADERSHIP_SUPPORT, LEADERSHIP_SUPPORT, 1)
]

DG = nx.DiGraph()
DG.add_weighted_edges_from(graph)
```

```python
centrality = nx.betweenness_centrality(DG, weight='weight')
sorted(centrality.items(),key=lambda x: x[1],reverse=True)
```

```python
rankings = nx.pagerank(DG)
```

```python
sorted(rankings.items(), key=lambda x: x[1], reverse=True)
```
